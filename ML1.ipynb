```python
import numpy as np
x=np.array([1, 3, 5, 7, 9])
y = np.array([75, 77, 85, 83, 90])
a = 2
b = 75
y_pred =[]
for i in range(len(x)):
    y_pred.append(a *x[i] +b)
def mse(y, y_pred):
    return (1/len(x)) *sum((y-y_pred)**2)
print('x:', x)
print('y:', y)
print('y_pred:', y_pred)
print('mse:', str(mse(y, y_pred)))
```

```python
import numpy as np
import matplotlib.pyplot as plt
x1= np.array([1, 3, 5, 7, 9])
x2= np.array([0, 1, 5, 2, 4])
y= np.array([75, 77, 85, 83, 90])
fig= plt.figure()
ax= fig.add_subplot(111, projection ='3d')
ax.scatter3D(x1, x2, y)
plt.title("Scatter plot")
plt.show()
a1= 0
a2= 0
b= 0
lr=0.0003 # learning rate
epochs = 20001 # epochs
n = len(x1) # number of x
for i in range(epochs):
  y_pred = a1 * x1 + a2 * x2+ b
  error = y-y_pred
  a1_diff = (2/n) * sum(-x1 * error)
  a2_diff = (2/n) * sum(-x2 * error)
  b_diff = (2/n) * sum(-error)
  a1= a1-lr* a1_diff
  a2= a2-lr* a2_diff
  b = b-lr * b_diff
  if(i % 2000 == 0):
    print('epoch: {}, a1: {}, a2: {}, b: {}'.format(i, a1, a2, b))
y_pred = a1 * x1 + a2 * x2 + b
print('y:', y)
print('y_pred:', y_pred)
```
